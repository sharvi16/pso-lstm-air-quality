```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\admin\OneDrive\Desktop\GDG\Task1.csv")
df.head()
```

```python
def crab_processing(df):
    # Drop the 'id' column
    df = df.drop('id', axis=1)
    
    # Drop all rows where 'Height' is 0
    df = df[df['Height'] != 0]
    
    # Create 'LostWeight' column
    weight_diff = df['Weight'] - (df['Shucked Weight'] + df['Viscera Weight'] + df['Shell Weight'])
    df['LostWeight'] = (weight_diff >= 0).astype(int)
    
    # One hot encoding for 'Sex' column
    df = pd.get_dummies(df, columns=['Sex'], prefix='Sex')
    
    return df
```

```python
df_crab = crab_processing(df.copy())
df_crab.head()
```

```python
df_crab.shape
```

```python
# Create X (features) and y (target)
X = df_crab.drop('Age', axis=1)
y = df_crab['Age']

print(f"X shape: {X.shape}")
print(f"y shape: {y.shape}")
print(f"\nX columns: {list(X.columns)}")
print(f"\nFirst few rows of X:")
print(X.head())
print(f"\nFirst few values of y:")
print(y.head())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
# Check the 0.3 ratio
total_samples = len(X)
test_samples = len(X_test)
train_samples = len(X_train)

test_ratio = test_samples / total_samples
train_ratio = train_samples / total_samples

print(f"Total samples: {total_samples}")
print(f"Training samples: {train_samples} ({train_ratio:.2%})")
print(f"Test samples: {test_samples} ({test_ratio:.2%})")
print(f"\nVerification: {train_samples} + {test_samples} = {train_samples + test_samples}")
```
